```python
import pandas as pd

train = pd.read_csv("./data/train.csv")
train_deflection_label = []

for index, row in train.iterrows():
  text = row['text'].lower()
  # train_deflection_label.append(sum([text.count(term)for term in deflection_language]))
  train_deflection_label.append(max([1 if term in text else 0 for term in deflection_language]))
print(sum(train_deflection_label))
train["deflection"] = train_deflection_label
train.to_csv('./data/train_deflection.csv')
```

```python
import pandas as pd

file_names = []
path = 'data/reports'
for file in os.listdir(path):
  if os.path.isfile(path + '/' + file):
    name = file.split('.')[0]
    file_names.append(name)

for file in file_names:
  test = pd.read_csv("./data/"+file+"-data.csv")
  test_deflection_label = []

  for index, row in test.iterrows():
    text = row['text'].lower()
    # test_deflection_label.append(sum([text.count(term)for term in deflection_language]))
    test_deflection_label.append(max([1 if term in text else 0 for term in deflection_language]))
  print(sum(test_deflection_label))
  test["deflection"] = test_deflection_label
  test.to_csv('./data/'+file+'_deflection.csv')
```
