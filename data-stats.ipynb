```python
def get_data_stats(df, col_list):
  print("Total number of examples: " + str(df.shape[0]))
  for col in col_list:
    print("Count for ", col, ": ", str(df[col].sum()))
  return
```

```python
import pandas as pd

train = pd.read_csv('data/FINAL-train-data.csv')
val = pd.read_csv('data/FINAL-val-data.csv')
```

```python
get_data_stats(train, ['binary_sentiment', 'specificity', 'commitment', 'hedging', 'label'])
```

```python
get_data_stats(val, ['binary_sentiment', 'specificity', 'commitment', 'hedging', 'label'])
```
